If you want to get the prior value, you can use For Loops. But you can also use the np where method and the shift method. As you can see at the bottom, using these methods together is much faster.

```python
import numpy as np
import pandas as pd
```

```python
#It took 8 minutes for a column, but it works.
#I don't know if doing None vs doing "" makes a difference in speed.
def prior_amount(curr_field_name, prior_field_name):
    a1 = Data.columns.get_loc('full_stmt_nbr')
    b1 = Data.columns.get_loc(curr_field_name)
    c1 = Data.columns.get_loc(prior_field_name)
    for x in range(1, Data_Row_Count):
        if Data.iloc[x,a1] == Data.iloc[x-1,a1]:
            Data.iloc[x,c1] = Data.iloc[x-1,b1]
        else:
            Data.iloc[x,c1] = None
```

```python
Data = pd.read_excel('test mpm type.xlsx', sheet_name = "Sheet1")
```

```python
# Store the row count of this data frame
Data_Row_Count = len(Data.index)
Data_Row_Count
```

```python
Data["Prior_Paid"] = np.where(Data.full_stmt_nbr == Data.full_stmt_nbr.shift(1), "Yes", "No")
```

```python
# np way
%time Data["Prior_Paid2"] = np.where(Data.full_stmt_nbr == Data.full_stmt_nbr.shift(1), Data.Paid.shift(1), None)
```

```python
# For Loop way
Data['Prior_Paid3'] = None
%time prior_amount(curr_field_name = 'Paid', prior_field_name = 'Prior_Paid3')
```
